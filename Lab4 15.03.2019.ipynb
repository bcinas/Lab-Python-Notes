```python
def main():
    select = 0
    while select != 5:
        print('---------------------------')
        print('To create a new file, press 1.')
        print('To append data to your file, press 2.')
        print('To display file data, press 3.')
        print('To search for a name, press 4.')
        print('To exist, press 5.')
        print('---------------------------')

        try:
            select = int(input('Enter your choice: '))

            if select == 1:
                writeFile()
            elif select == 2:
                appendFile()
            elif select == 3:
                readFile()
            elif select == 4:
                findName()
            elif select == 5:
                message()
            else:
                print('Retry!')

        except ValueError as err:
            print(err)

def findName():
    try:
        infile = open('Names.txt','r')
        name = input('Enter the name you are looking for ')
        found = False
        line = infile.readline()

        while line != '':
            if name == line.rstrip():
                found = True
            line = infile.readline()
        if found:
            print(name, 'is available in the file.')
        else:
            print(name, 'is not available in the file.')

    except Exception as err:
        print(err)

def message():
    print('You have request to quit.')
    print('--------> END <----------')

def writeFile():
    outfile = open('Names.txt', 'w')

    for x in range(7):
        name = input('Enter a city name: ')
        outfile.write(name + '\n')

    outfile.close()

def readFile():
    infile = open('Names.txt', 'r')
    line = infile.readline()

    while line != '':
        print(line.rstrip())
        line = infile.readline()

    infile.close()

def appendFile():
    outfile = open('Names.txt', 'a')
    num = int(input('How many names would you like to add? '))

    for x in range(num):
        outfile.write(input('Enter a city name: ')+'\n')

    outfile.close()
main()
```
